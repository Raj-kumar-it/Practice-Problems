```python
##For sales dataset do a Time Series visualization
#1)_here i am importing sales data set 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('sales1.csv',  index_col="DATE")
df
```

```python
#INDEX:It is a default index type that can be used for DataFrame or Series when there is no explicit index 
#which is provided by the user.
#By using rangeindex we can improve computing speed.
df.index
```

```python
# is null() a method returns a DataFrame object where all the values are replaced with a Boolean value True For NULL values
#otherwise False
df.isnull()
```

```python
# not null() function detects existing/non-missing values in the DataFrame
df.notnull()
```

```python
#Dropna()=Remove all rows wit NULL values from the DataFrame
df.dropna()
```

```python
#The fillna() method replaces the NULL values with a specified value. 
#The fillna() method returns a new DataFrame object unless the inplace parameter is set to True
df.fillna(0)
```

```python
#Differencing:
#Differencing is used to make the difference in values of a specified interval
df.TV.diff(2).plot(figsize=(10, 6))
```

```python
df.Radio.diff(2).plot(figsize=(10, 5))
```

```python
df.Sales.diff(2).plot(figsize=(10, 6))
```

```python
#Shift():
#The shift function can be used to shift the data before or after the specified time interval
#We can specify the time, and it will shift the data by one day by default
#It is helpful to see previous day data and todayâ€™s data simultaneously side by side.
df['Change'] = df.Sales.div(df.Sales.shift())
df['Change'].plot(figsize=(10, 8), fontsize=16)
```

```python
##TV with shift function
df['Change'] = df.TV.div(df.TV.shift())
df['Change'].plot(figsize=(10, 8), fontsize=16)
```

```python
#We can also take a specific interval of time and plot to have a clearer look
#Here we are plotting the data of only 2006-09.
df['2006-09']['Change'].plot(figsize=(10, 6))
```
