## visualizing Geospatial data in python using Folium Library

## Here we import required Libraries
-folium
-ipywidgets
-geocoder
-geopy
-numpy
-pandas


```python
import folium
import ipywidgets
from folium import plugins
import geocoder
import geopy
import pandas as pd
import numpy as np
```

# displaying a simple GIS map


```python
map=folium.Map(location=[20.5937,78.9629],tiles="Stamen Terrain",zoom_start=5)
map.add_child(folium.LatLngPopup()) # we can see the lantitude and langitude for any loaction
map
```

```python

map=folium.Map(location=[20.5937,78.9629],tiles="Stamen Terrain",zoom_start=5)
measure_control=plugins.MeasureControl(position="topleft",
                                       active_color="red",
                                       completed_color="red",
                                       primary_length_units="meters")
map.add_child(measure_control)
map
```

```python
usa=pd.read_csv("usa geospatial.csv")
usa
```

# lets display 10 records from this file

# display all the wells on the map

```python
map=folium.Map(location=[29.53,-99.11],zoom_start=5)
usa.apply(lambda row:folium.CircleMarker(location=[row["LATITUDE"],row["LANGITUDE"]],
                                        radius=10).add_to(map),axis=1)
map
```
