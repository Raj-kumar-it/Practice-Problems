```python
#1)Create different Bar plots for variables in any dataset
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('traintic.csv')
df
```

```python
#some preprocessers or built methods to handle the in data set
#Head() Functio is used to get the first n rows 
df.head()
```

```python
#Tail() Function is used to get the last n rows
df.tail()
```

```python
# is null() a method returns a DataFrame object where all the values are replaced with a Boolean value True For NULL values
#otherwise False
df.isnull()
```

```python
# not null() function detects existing/non-missing values in the DataFrame
df.notnull()
```

```python
#shape() method is used to fetch the dimensions of pandas and numpy type objects in python
df.shape
```

```python
#To Know About DataSet
df.info
```

```python
#to display sepcific rows using slicing
df.iloc[5:15]
```

```python
df.iloc[4:-1]
```

```python
#INDEX:It is a default index type that can be used for DataFrame or Series when there is no explicit index 
#which is provided by the user.
#By using rangeindex we can improve computing speed.
df.index
```

```python
#Dropna()=Remove all rows wit NULL values from the DataFrame
df.dropna()
```

```python
#The fillna() method replaces the NULL values with a specified value. 
#The fillna() method returns a new DataFrame object unless the inplace parameter is set to True
df.fillna(4)
```

```python
#to change columns name in data set
df.rename(columns={'Name':'NAME','Ticket':'TICKET'},inplace=True)
df
```

```python
plt.bar(df['SEX'],df['Age'])
plt.title("Bar Plot with grid lines")
plt.xlabel('Sex')
plt.ylabel('Age')
plt.grid()
plt.show()
```

```python
plt.barh(df['Sex'],df['Age'])
plt.title("horizantal Bar Plot with grid lines")
plt.xlabel('Sex')
plt.ylabel('Age')
plt.grid()
plt.show()
```

```python
plt.bar(df['Age'],df['Fare'])
plt.title("Bar Plot for ages")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
plt.bar(df['Survived'],df['Pclass'])
plt.title('BAR CHART')
plt.xlabel('Survived')
plt.ylabel('Pclass')
plt.show()
```

```python
plt.bar(df['Sex'],df['Pclass'],color='red',width=0.5)
plt.title('BAR CHART WITH COLOR')
plt.xlabel('Sex')
plt.ylabel('Pclass')
plt.show()
```

```python
plt.bar(df['Pclass'],df['Age'],color='black',width=0.5)
plt.title('BAR CHART WITH COLORS AGE VS PCLASS')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.show()
```
