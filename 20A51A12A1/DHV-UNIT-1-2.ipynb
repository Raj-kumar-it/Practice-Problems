```python
#Dfferent Color Graphs
#1)Bar Graph
plt.bar(dv1['JAN'],dv1['FEB'])
plt.title("Bar Chart")
plt.xlabel('JAN')
plt.ylabel('FEB')
plt.show()
```

```python
##HISTOGRAM FOR ANNUAL(attribute)
plt.hist(dv1['ANNUAL'])
plt.title("HISTOGRAM")
plt.show()
```

```python
##HISTOGRAM FOR JAN(attribute)
plt.hist(dv1['JAN'])
plt.title("HISTOGRAM")
plt.show()
```

```python
##SCATTER PLOT for  and 
plt.scatter(dv1['ANNUAL'],dv1['YEAR'],c=dv1['JAN'],s=dv1['Jan-Feb'])
plt.title("SCATTER PLOT")
plt.xlabel('ANNUAL')
plt.ylabel('YEAR')
plt.colorbar()
plt.show()
```

```python
plt.plot(dv1['YEAR'])
plt.plot(dv1['ANNUAL'])
plt.title("SCATTER PLOT")
plt.xlabel('ANNUAL')
plt.ylabel('YEAR')
plt.show()
```

```python
##HISTOGRAM for ANNUAL
dv1.ANNUAL.hist(bins=10)
```

```python
##HISTOGRAM for all attributes
dv1.hist(bins=10)
```

```python
##DENSITY plot for ANNUAL
dv1.ANNUAL.plot.density(color='red')
plt.show()
```

```python
##DENSITY plot for YEAR
dv1.YEAR.plot.density(color='green')
plt.show()
```

```python
##DENSITY plot for MAR
dv1.MAR.plot.density(color='yellow')
plt.show()
```

```python
##MULTIPLE DENSITY PLOT
data_wide=dv1.pivot(columns='YEAR',values='ANNUAL')
data_wide.head()
data_wide.plot.density(figsize=(7,7),linewidth=4)
plt.xlabel("temperatures")
```
